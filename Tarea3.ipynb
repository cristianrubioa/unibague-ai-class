# 4. ¿Es posible usar una red neuronal para este caso?, ¿Se debe mejorar el sistema o el tipo de datos ingresados a la red?

Claro que es posible entrenar la red neuronal, es más, estoy a punto de desarrollar la red y entrenarla en la linea posterior en 3, 2, 1... 😎

```python
# Preparamos nuestra Red adaline

class AdalineGD(object):
    """ADAptive LInear NEuron classifier.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.


    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    cost_ : list
      Sum-of-squares cost function value in each epoch.

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        """ Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
          Training vectors, where n_samples is the number of samples and
          n_features is the number of features.
        y : array-like, shape = [n_samples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        self.cost_ = []

        for i in range(self.n_iter):
            net_input = self.net_input(X)
            # Please note that the "activation" method has no effect
            # in the code since it is simply an identity function. We
            # could write `output = self.net_input(X)` directly instead.
            # The purpose of the activation is more conceptual, i.e.,  
            # in the case of logistic regression (as we will see later), 
            # we could change it to
            # a sigmoid function to implement a logistic regression classifier.
            output = self.activation(net_input)
            errors = (y - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self
        

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        """Compute linear activation"""
        return X

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.activation(self.net_input(X)) >= 0.0, 1, -1)
```

## 4.1. Entrenamiento & Predicción (opción A)

```python
# Llamamos a la clse AdalineGD() y ejecutamos el metodo fit()
# para entranar la red Adaline con los datos de la opción A.
adaline = AdalineGD()
adaline.fit(Xa, ya)

# Datos predecidos
print("Datos predecidos:")
print(adaline.predict(Xa))
print("\n")

# Datos reales
print("Datos reales: ")
print(ya)
print("\n")

print("Parece bien 👍🏼 Clasificó correcctamente")
print("Somos capaces de predecir nuestro datset ficticio 🥉")
```

Lo que vamos a hacer para este ejemplo es seguir el orden denotado en la tabla, es decir, primero vamos a entrenar nuestro modelo dados C2 y C3 para la clase o salida correspondiente de (C2 OR C3) y luego (C1 AND (C2 or C3)), para esto tomamos la predicción que arroja el clasificador anterior y lo ingresamos como entrada de nuestro nuevo modelo para entrenarlo. "Sip, vamos a hacer dos clasificadores", podríamos decir que nuestro primer clasificador pertenece a una capa oculta en terminos generales aunque no es una definiciónn correcta, puesto que al final lo que queremos es obtener el resultado final y no me importa si tuve que pasar por pasos intermedios (cabe resaltar que el algoritmo no es automático).
```python
# Trazamos la función de decisión 

from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):

    # Generador de marcadores de configuración y mapa de colores
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # Trazar la superficie de decisión
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # Muestras de clase de trazado
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
```
